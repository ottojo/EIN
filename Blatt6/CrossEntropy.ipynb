```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bob = np.array([1/2,1/4,1/8,1/8])

tdelta = 0.01
tValues = np.arange(0.01,1,tdelta)
```

```python
def c(t): # t shall be scalar
    return np.array([(3/4) * t, (1-t)*(3/4), 1/8, 1/8])
```

```python
def H(q,p):
    return np.dot(p, - np.log2(q))
```

```python
def KLdivergence(q,p):
    return H(q,p) - H(p,p)
```

```python
def calcDivergences():
    d = []
    for t in tValues:
        d.append(KLdivergence(c(t),bob))
    return d
KLdivergences = calcDivergences()
```

```python
fig= plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_title('Changing the code of Charles to fit to Bob \'s distribution')
ax1.set_xlabel('t')
ax1.set_ylabel('$D_C(B)$')
ax1.grid()

ax1.plot(tValues, KLdivergences, label='KL-Divergence')
ax1.legend()
```
